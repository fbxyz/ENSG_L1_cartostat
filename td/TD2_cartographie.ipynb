## 2.2 Données attributaires
Il n'est pas conseillé de faire glisser le fichier CSV dans Qgis. Les délimiteurs (;) et décimales (,) n'étant pas standards, la table sera mal chargée

Allez dans Couche &#8594; Ajouter une couche &#8594; Ajouter une couche de texte délimité

<img src="https://i.imgur.com/ehJfXRy.png" style="margin-left:auto; margin-right:auto"/>


Puis renseignez la fenêtre avec les paramètres suivants :

- **Format de Fichier** --> délimiteurs personnalisés virgules (pensez à décocher les autres)
- **Option des champs et enregistrements** --> Virgule en séparateur point doit être **décoché**, **en-têtes en 1ère ligne** et **Détecter les types de champs** doivent être cochés.
- **Définition de la géométrie** --> pas de géométrie

Dans Echantillon de données, vérifiez que la table est bien formatée : 
- **Code** et **Département** en Texte
- __ICF*__ en Décimal (double)
  
Enfin, ajoutez la table
